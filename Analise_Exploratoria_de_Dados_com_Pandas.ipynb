```python
#Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
df = pd.read_excel('AdventureWorks.xlsx')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
# formatandopara que as casas decimais sejam limitadas a duas casas
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# verificando receita total

df["Valor Venda"].sum()
```

```python
# criando uma coluna do custo total (custo unitário * quantidade)

df['Custo'] = df['Custo Unitário'] * df['Quantidade']
df.head()
```

```python
# agora analisando qual o custo total
df['Custo'].sum()
```

```python
# Criando coluna de Lucro = Receita - Custo

df['Lucro'] = df['Valor Venda'] - df['Custo']
df.head()
```

```python
# Lucro total
round(df['Lucro'].sum(),2)
```

```python
# lucro médio por operação

lucro_medio = df['Lucro'].sum() / df['Lucro'].count()
lucro_medio
```

```python
# analisando tempo de envio
df['Tempo_Envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
df.head()
```

```python
#Média do tempo de envio por Marca
df.groupby("Marca")["Tempo_Envio"].mean()
```

```python
#Verificando se temos dados faltantes
df.isnull().sum()
```

```python
#Vamos Agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#Resetando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita");
```

```python
#Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
df_2009.head()
```

```python
#verificando as principais caracteristicas estatísticas de tempo de envio 
df["Tempo_Envio"].describe()
```
